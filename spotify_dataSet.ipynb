```python
import pandas as pd
```

```python
df= pd.read_csv('/content/spotify_data clean.csv')
```

```python
print("The Information of Data is:")
df.info()
```

```python
print("The Description of Data is:")
df.describe()
```

```python
print("First Five Rows of Data is:")
df.head()
```

```python
print("After Date/ Time Conversion:")
df['album_release_date']= pd.to_datetime(df['album_release_date'])
df.info()
print("Converted column head")
print(df['album_release_date'].head())
```

```python
print("Handles Duplicates:")
df.drop_duplicates(subset=['track_name', 'artist_name'], keep= 'first', inplace= True)
df.head()
```

```python
print("Extracted Year:")
df['release_year']= df['album_release_date'].dt.year
print("Extracted Month:")
df['release_month']= df['album_release_date'].dt.month

df.head()
```

```python
df.info()
```

```python
artists_count= df.groupby('artist_name').size().sort_values(ascending=False)
```

```python
df['artist_name']= df['album_name'].str.lower().str.strip()
```

```python
print(artists_count.head(10))
```

```python
df.dropna(subset=['artist_name'], inplace= True)
```

```python
print(artists_count.head(10))
```

```python
df.info()
```

```python
df_filtered= df[df['artist_name'].str.len() < 30]
print(f"Original Rows: {len(df)}. Filtered Rows: {len(df_filtered)}")
```

```python
top_artists = df_filtered.groupby('artist_name').size().sort_values(ascending=False)
```
